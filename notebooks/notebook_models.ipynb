```python
import pandas as pd
linreg = pd.read_csv('data/error_metrics/error_linear_reg.csv')
polreg = pd.read_csv('data/error_metrics/error_pol_reg3.csv')
rf = pd.read_csv('data/error_metrics/error_randomforest.csv')
nn = pd.read_csv('data/error_metrics/error_neuralnetwork.csv')
```

```python
linreg = linreg['Error']
polreg = polreg['Error']
rf = rf['Error']
nn = nn['Error']
```

```python
mae = [linreg[0], polreg[0], rf[0], nn[0]]
mape = [linreg[1], polreg[1], rf[1], nn[1]]
mse = [linreg[2], polreg[2], rf[2], nn[2]]
rmse = [linreg[3], polreg[3], rf[3], nn[3]]
```

```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Valores de X
x_values = np.arange(4)

# Ancho de cada barra
bar_width = 0.2

# Crear gráfico de barras
plt.bar(x_values, linreg, color='#62B365', width=bar_width, label='Linear Reg')
plt.bar(x_values + bar_width, polreg, color='#4378BD', width=bar_width, label='Poly Reg (3)')
plt.bar(x_values + 2*bar_width, rf, color='#473982', width=bar_width, label='Random Forest')
plt.bar(x_values + 3*bar_width, nn, color='#D55369', width=bar_width, label='Neural Network')

# Configurar ejes y leyenda
plt.xticks(x_values + 1.5*bar_width, ['MAE', 'MAPE', 'MSE', 'RMSE'])
plt.ylabel('Valores')
plt.legend()
plt.title('Métricas de error de algunos modelos')

plt.savefig('resources/img/error_metrics.png')

```
